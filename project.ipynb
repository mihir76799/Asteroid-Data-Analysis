```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
data = "NASA_Close_Approach_Data_1960_2060.csv"
df=pd.read_csv(data)
df["dist"]=pd.to_numeric(df["dist"], errors="coerce")
df["v_rel"]=pd.to_numeric(df["v_rel"], errors="coerce")
df=df.dropna()

df["dist"].fillna(df["dist"].mean())
df["v_rel"].fillna(df["v_rel"].mean())

sns.histplot(df["dist"])
plt.title("Distribution of asteroid close approach distances")
plt.xlabel("Distance (AU)")
plt.ylabel("Number of asteroids")
plt.show()

sns.histplot(df["v_rel"])
plt.title("Relative Velocity of asteroids")
plt.xlabel("Velocity of asteroids (km/s)")
plt.show()

plt.figure(figsize=(6,6))
plt.scatter(df["dist"],df["v_rel"], color="black")
plt.title("Relative velocity vs close approach distances")
plt.ylabel("Relative velocity of asteroids")
plt.xlabel("Asteroid close approach distances")
plt.show()
```
